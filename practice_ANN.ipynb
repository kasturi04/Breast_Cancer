```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#drop id and unnamed column
df = df.drop(columns = ['id','Unnamed: 32'])
df.head()
```

```python
# change the diagnosis to binary values
df.loc[df['diagnosis']=='M','benign_0__mal_1']=1
df.loc[df['diagnosis']=='B','benign_0__mal_1']=0
df.head()
```

```python
df=df.drop(columns=['diagnosis'])
df.head()
```

```python
sns.countplot(x='benign_0__mal_1',data=df)
plt.show()
```

```python
x=df.drop(columns=['benign_0__mal_1'])
y=df['benign_0__mal_1']
x.head()
```

```python
#split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#scaling the data
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#create ann with tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
```

```python
#create ann 
model=Sequential()
model.add(InputLayer(input_shape=(30,)))
model.add(Dense(26,activation='relu'))      
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#train the model
model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```

```python
#PLOT LOSS AND ACCURACY GRAPHS
import matplotlib.pyplot as plt
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
plt.show()
```

```python
#EARLY STOPPING THE MODEL
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=20)
model.fit(x_train, y_train, epochs=500, validation_data=(x_test, y_test), callbacks=[early_stop])
```

```python
import pickle
model.save("BC_model.h5")

with open("BC_scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```
